## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("diamonds")
data.head()
```

```
# Carregar o conjunto de dados 'diamonds' do seaborn
diamonds = sns.load_dataset('diamonds')
```

```
# gráfico do exercício 1

# Criar um gráfico de barras para mostrar o preço médio dos diamantes por tipo de corte
plt.figure(figsize=(10, 6))
sns.barplot(x='cut', y='price', data=diamonds, estimator=lambda x: sum(x)/len(x), palette='viridis')

# Adicionar título e rótulos aos eixos
plt.title('Preço Médio dos Diamantes por Tipo de Corte')
plt.xlabel('Tipo de Corte')
plt.ylabel('Preço Médio (USD)')
plt.show()
```

**Insight do gráfico 1**: Os tipos de diamantes "Premium" e "Fair" são os maiores preços médios
Isso sugere que a qualidade do corte influencia significativamente o preço dos diamantes. Investir em cortes de maior qualidade pode aumentar o valor de mercado dos diamantes, pois cortes superiores são mais valorizados pelos consumidores.

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("flights")
data.head()
```

```
# Filtrar os dados para incluir apenas os registros de dezembro
december_data = data[data['month'] == 'Dec']
```

```
# gráfico do exercício 2

# Criar um gráfico de linhas para mostrar o número de passageiros em dezembro por ano
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='passengers', data=december_data, marker='o', color='b')

# Adicionar título e rótulos aos eixos
plt.title('Número de Passageiros em Dezembro por Ano')
plt.xlabel('Ano')
plt.ylabel('Número de Passageiros')
plt.grid(True)
plt.show()
```


**Insight do gráfico 2**: O gráfico mostra que o número de paasageiros em dezembro vem crescendo mais a cada ano, pórem teve uma leve lateralização em 1957 para 1958 voltando a crescer no período seguinte

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3

# Criar um gráfico de linhas para mostrar o número de passageiros por mês entre 1949 e 1959
plt.figure(figsize=(14, 8))
sns.lineplot(x='year', y='passengers', hue='month', data=data, palette='tab10', marker='o')

# Adicionar título e rótulos aos eixos
plt.title('Número de Passageiros por Mês entre 1949 e 1959')
plt.xlabel('Ano')
plt.ylabel('Número de Passageiros')
plt.legend(title='Mês', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

**Insight do gráfico 3**: Tendência de Crescimento: Há uma clara tendência de crescimento no número de passageiros ao longo dos anos. Isso indica um aumento na popularidade e acessibilidade das viagens aéreas durante esse período.
